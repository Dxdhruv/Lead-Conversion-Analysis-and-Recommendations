```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
leads_df = pd.read_csv('/content/lead_data.csv')
```

```python
leads_df
```

```python
location_counts = leads_df['Location'].value_counts()
```

```python
location_counts
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=location_counts.index, y=location_counts.values, palette="viridis")
plt.title('Number of Leads by location')
plt.xlabel('Location')
plt.ylabel('Number of Leads')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
program_counts = leads_df['Program Interest'].value_counts()
program_counts
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=program_counts.index, y=program_counts.values, palette="viridis")
plt.title('Number of leads by program interest', fontsize=16)
plt.xlabel('Program Interest', fontsize=12)
plt.ylabel('Number of leads', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
source_counts = leads_df['Lead Source'].value_counts()
plt.figure(figsize=(10,6))
sns.barplot(x=source_counts.index, y=source_counts.values, palette="viridis")
plt.xlabel('Lead Sources')
plt.ylabel('Number of Leads')
plt.title('Number of Leads sources')
plt.show()
```

```python
location_program_counts = leads_df.groupby(['Location', 'Program Interest']).size().unstack()
```

```python
location_program_counts
```

```python
location_program_counts.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Number of Leads by Location and Program Interest')
plt.xlabel('Location', fontsize=12)
plt.ylabel('Number of Leads', fontsize=12)
# plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
location_program_counts_long = location_program_counts.reset_index().melt(id_vars='Location', var_name='Program Interest', value_name='Count')
plt.figure(figsize=(12,8))
sns.barplot(data=location_program_counts_long, x='Location', y='Count', hue='Program Interest', palette='viridis')
plt.title('Program Interest by Location')
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
year_program_counts = leads_df.groupby(['Year of Study', 'Program Interest']).size().unstack()
```

```python
year_program_counts
```

```python
program_interest_percentage = leads_df['Program Interest'].value_counts(normalize=True) * 100
program_interest_percentage
```

```python
program_interest_percentage.plot(kind='bar', figsize=(10,6))
plt.title('Percentage Distribution of Leads by E-learning Program')
plt.xlabel('E-Learning Program')
plt.ylabel('Percentage by Leads')
plt.show()
```

```python
top_colleges = leads_df['College'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_colleges.values, y=top_colleges.index, palette="magma")
plt.title('Top Colleges Contributing Leads', fontsize=16)
plt.xlabel('Number of Leads', fontsize=12)
plt.ylabel('College', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
location_counts = leads_df['Location'].value_counts().head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=location_counts.index, y=location_counts.values, palette="viridis")
plt.title('Location Wise Lead Distribution')
plt.ylabel('Number of Leads')
plt.xlabel('Location')
plt.tight_layout()
plt.show()
```
